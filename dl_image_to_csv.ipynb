```python
import os

#get all path of all images in folder image_input

path = 'pizzacam/image_input'
files = []
for r, d, f in os.walk(path):
    for file in f:
        if '.jpg' in file:
            files.append(os.path.join(r, file))

# print(files)
```

```python
import os

_id = os.listdir('image_input/')
```

```python
# remove .jpg
id = []
for i in _id:
    id.append(i[:-4])
```

```python
new_df = data[data['_id'].isin(id)]
```

```python
new_df
```

```python
# suffle data and just take 12000 records

new_df = new_df.sample(frac=1).reset_index(drop=True)

new_df = new_df[:12000]

new_df
```

```python
# splitting data into train, validation and test

train_df = new_df.sample(frac=5/6, random_state=0)
test_df = new_df.drop(train_df.index)

```

```python
train_df.to_csv('pizzacam/train.csv', index=False)
test_df.to_csv('pizzacam/test.csv', index=False)
```
