```python
## Read json data
import pandas as pd
import json
# from __init__ import *

with open('../data/pizzacam.json') as f:
    data_json = json.load(f) # data is a list of dictionaries
```

```python
# Check label of data

pizza_error_list_key = set()
for i in range(len(data_json)):
    # if (i<830):
        entry_error_list = []
        for key in data_json[i].keys():
            if (key == 'error_list'):
                for error in data_json[i]['error_list']:
                    pizza_error_list_key.add(error)
```

```python
pizza_error_list_key = list(pizza_error_list_key)
```

```python
pizza_error_list_key
```

```python
pizza_error_list = list()
for i in range (len(data_json)):
    # if (i<830):
    entry_error_list = {
                        "_id" : data_json[i]['_id']['$oid'], 
                        "image_url" : data_json[i]['image_url']
                        }
    for key in data_json[i].keys():
        if (key == 'error_list'):
            for error in data_json[i]['error_list']:
                entry_error_list[error] =  data_json[i]['error_list'][error]

    pizza_error_list.append(entry_error_list)
```

```python

# save pizza_error_list to json file in foler data, an entry is a dictionary, and each entry is a line. In the end of line, write an ,
# Use UTF 8

with open('../data/pizza_error_list.json', 'w', encoding='utf-8') as f:
    for item in pizza_error_list:
        json.dump(item, f, ensure_ascii=False)
        f.write(',\n')
```

```python
def create_new_column_name(value):
    return value.replace(' | ', '_').replace(' ', '')
```

```python
def create_new_column_name_to_one_hot_vector(dataframe, column_name):
    for index, row in df.iterrows():
        if isinstance(row[column_name], list):
            for value in row[column_name]:
                col_name = create_new_column_name(value)
                if col_name not in df.columns:
                    df[col_name] = 0
                df.at[index, col_name] = 1
    return dataframe
```

```python
columns = ['Baking', 'Fermentation_condition', 'Edge', 'Topping', 'Advanced_Edge', 'Shape', 'Size']

for column in columns:
    df = create_new_column_name_to_one_hot_vector(df, column)

df.head()
```
